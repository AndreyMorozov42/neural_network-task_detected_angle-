```python
import matplotlib.pyplot
import numpy as np
import csv
import os

# описание путей
path = "C:\\Users\\andmo\\OneDrive\\Desktop\\task_tuesday\\train-set"
name_file = "train-set.csv"

# меняет текущую директорию
os.chdir(path)

with open(name_file, "r", encoding="utf-8") as file_open:
    # считываем данные из файла
    data = list(csv.reader(file_open, delimiter = ","))

# заполняем массив правильных ответов
goal_set = []
for i in range(len(data)):
    goal = [0, 0, 0, 0]
    if int(data[i][0]) == 1:
        goal[int(data[i][0]) - 1] = 1
    elif int(data[i][0]) == 2:
        goal[int(data[i][0]) - 1] = 1
    elif int(data[i][0]) == 3:
        goal[int(data[i][0]) - 1] = 1
    elif int(data[i][0]) == 4:
        goal[int(data[i][0]) - 1] = 1
    goal_set.append(goal)

goal_set = np.array(goal_set)
```

```python
# количество проходов по учебному набору
iteration = 350
# коэф. обучения
alpha = 0.01
# инициация размера массива
num_pixels = 900
# инициация скрытого слоя
num_hide = 140
# инициация размера выхода
num_output = 4

relu = lambda x: (x>=0) * x
relu2deriv = lambda x: x>=0

# объявление весов
weight_0_1 = 0.2*np.random.random((num_pixels, num_hide)) - 0.1
#print(weight_0_1)
weight_1_2 = 0.2*np.random.random((num_hide, num_output)) - 0.1
#print(weight_1_2)

for j in range(iteration):
    error = 0
    for i in range(len(data)):
        # получение входных значений для первого слоя
        layer_0 = np.array(data[i][1:])
        layer_0 = layer_0.astype(float)
        
        # прямое распространение 
        layer_1 = relu(np.dot(layer_0, weight_0_1))
        layer_2 = np.dot(layer_1, weight_1_2)
        
        # ошибка
        error += np.sum((layer_2 - goal_set[i]) ** 2)
        
        # обратное распространение
        layer_2_delta = (goal_set[i] - layer_2)
        layer_1_delta = np.dot(layer_2_delta, weight_1_2.T) * relu2deriv(layer_1)
        
        weight_1_2 += alpha * layer_1.reshape(-1,1) * layer_2_delta
        weight_0_1 += alpha * layer_0.reshape(-1,1) * layer_1_delta
    print(f'I:{j} Error:{error}')
```

```python
# проверка эффективности нейронной сети
eff = []

# описание путей
path = "C:\\Users\\andmo\\OneDrive\\Desktop\\task_tuesday\\test-set"
name_file = "test-set.csv"

# меняет текущую директорию
os.chdir(path)

# открытие файла
file_open = open(name_file, "r", encoding="utf-8")

# считываем данные из файла
data = list(csv.reader(file_open, delimiter = ","))

file_open.close()

for i in range(len(data)):
        # получение входных значений для первого слоя
        layer_0 = np.array(data[i][1:])
        layer_0 = layer_0.astype(float)
        
        # прямое распространение 
        layer_1 = relu(np.dot(layer_0, weight_0_1))
        layer_2 = np.dot(layer_1, weight_1_2)
        
        # индекс ассоциируется с типом угла
        ind = list(layer_2).index(max(layer_2))
        
        if int(ind) == int(data[i][0]) - 1:
            eff.append(1)
        else:
            eff.append(0)
        
        
print(f'Итоговая эффективность на тестовом наборе: {eff.count(1) / len(eff) * 100 } %')
print(eff)
```
